```python
current_url = driver.current_url
```

```python
Headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'}
data = requests.get(current_url, headers=Headers)
soup = BeautifulSoup(data.text, "html.parser")

Title = []
Year = []
Duration = []
Rating = []

list_items = soup.find("ul", class_="ipc-metadata-list ipc-metadata-list--dividers-between sc-315eec58-0 hvFmRn detailed-list-view ipc-metadata-list--base").find_all('li')
for res in list_items:
    try:
        Title.append(res.find('h3').get_text().split('.')[1])
    except:
        Title.append('')
        
    try:
        Year.append(res.find('span',class_='sc-43986a27-8 jHYIIK dli-title-metadata-item').get_text())
    except:
        Year.append('')
        
    try:
        Duration.append(res.find('div',class_='sc-43986a27-7 dBkaPT dli-title-metadata').get_text().split(year)[1].split('m')[0]+'m')
    except:
        Duration.append('')
    
    try:
        Rating.append(res.find('span',class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating').get_text())
    except:
        Rating.append('')
    
    output = {'Title':Title, 'Year':Year, 'Duration': Duration, 'Rating':Rating}
```

```python
imdb_df = pd.DataFrame(output)
imdb_df.to_excel("IMDB_Automation.xlsx",index=False)
```
